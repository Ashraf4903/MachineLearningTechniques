# Define the Custom Perceptron Class

```python
# This is the custom Single-Layer Perceptron class exactly as defined in your lab manual
class Perceptron:
    def __init__(self, learning_rate=0.01, epochs=100):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None


    def activation(self, x):
        return np.where(x >= 0, 1, 0)

    # Training function
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        for _ in range(self.epochs):
            for i in range(n_samples):
                # Calculate the dot product of inputs and weights
                linear_output = np.dot(X[i], self.weights) + self.bias
                y_pred = self.activation(linear_output)
                
                # Update weights and bias based on the error
                update = self.lr * (y[i] - y_pred)
                self.weights += update * X[i]
                self.bias += update

    # Prediction function
    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return self.activation(linear_output)
        
print("Custom Perceptron class successfully defined!")
```

# Train and Evaluate the Model

```python

model = Perceptron(learning_rate=0.01, epochs=1000)

print("Training the custom Perceptron... (this might take a few seconds)")
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
